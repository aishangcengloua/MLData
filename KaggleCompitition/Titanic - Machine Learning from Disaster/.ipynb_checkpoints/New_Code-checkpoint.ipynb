```python
import pandas as pd#处理表格数据
import matplotlib.pyplot as plt
train_data = pd.read_csv('train.csv')
#train_data.head()
#train_data.tail()
#DataFrame为制表，字典中的key是column
fig = plt.figure(figsize = (15, 5))
fig.set(alpha = 0.2)

plt.subplot2grid((1, 3), (0, 0))
#在一张大图里分列几个小图，（2，3）代表将整个画板化为2行3列，（0，0）表示图在画板上的位置
train_data["Survived"].value_counts().plot(kind = 'bar')#df.plot(kind = "")pandas作图
plt.title("Rescue situation (1 is survived)")
plt.ylabel("number of person")

plt.subplot2grid((1, 3), (0, 1))
train_data["Pclass"].value_counts().plot(kind = 'bar')
plt.title("Pclass distribution of passengers")
plt.ylabel("number of person")

plt.subplot2grid((1, 3), (0, 2))
train_data["Embarked"].value_counts().plot(kind = 'bar')
plt.title("each embarked board passengers")
plt.ylabel("number of person")
plt.show()

fig = plt.figure()
fig.set(alpha = 0.2)#画板的透明度
plt.scatter(train_data["Survived"], train_data["Age"])
plt.grid(b = True, which = 'major', axis = 'y')
plt.title("Survived distribution by age (1 is survived)")
plt.ylabel("age")

fig = plt.figure()
fig.set(alpha = 0.2)
train_data["Age"][train_data["Pclass"] == 1].plot(kind = 'kde')#密度图
train_data["Age"][train_data["Pclass"] == 2].plot(kind = 'kde')
train_data["Age"][train_data["Pclass"] == 3].plot(kind = 'kde')
plt.title("Pclass of passengers by age")
plt.xlabel("age")
plt.ylabel("density")
plt.legend(("first class", "second class", "third class"), loc = 'best')#图例的位置

fig = plt.figure()
fig.set(alpha = 0.2)
survived_0 = train_data["Pclass"][train_data["Survived"] == 0].value_counts()
survived_1 = train_data["Pclass"][train_data["Survived"] == 1].value_counts()
df = pd.DataFrame({'survived' : survived_1, 'not survived' : survived_0})
df.plot(kind = 'bar', stacked = True)
plt.title("survived passengers distribution by Pclass")
plt.xlabel("Pclass")
plt.ylabel("number of person")
plt.show()

fig = plt.figure()
fig.set(alpha =0.2)

survived_m = train_data["Survived"][train_data["Sex"] == 'male'].value_counts()
survived_f = train_data["Survived"][train_data["Sex"] == 'female'].value_counts()
df = pd.DataFrame({'male' : survived_m, 'female' : survived_f}).plot(kind = 'bar', stacked = True)
plt.title("survived passengers distribution by Sex")
plt.xlabel("Sex")
plt.ylabel("number of person")
plt.show()

fig = plt.figure(figsize=(15, 5))
fig.set(alpha = 0.65)
plt.title("survived passengers distribution by Pclass and Sex")

ax1 = fig.add_subplot(141)#114的意思是将画板分成1 * 4的规格，后面的1表示第一个部分
train_data["Survived"][train_data["Sex"] == 'female'][train_data["Pclass"] != 3].value_counts().plot(kind = 'bar', color = '#FA2479')
ax1.set_xticklabels(['survived', "not survived"], rotation = 0)
plt.legend(['female / low class'], loc = 'best')

ax2 = fig.add_subplot(142, sharey = ax1)
train_data["Survived"][train_data["Sex"] == 'female'][train_data["Pclass"] == 3].value_counts().plot(kind = 'bar', color = 'pink')
ax2.set_xticklabels(['survived', "not survived"], rotation = 0)
plt.legend(['female / high class'], loc = 'best')

ax3 = fig.add_subplot(143, sharey = ax1)
train_data["Survived"][train_data["Sex"] == 'male'][train_data["Pclass"] != 3].value_counts().plot(kind = 'bar', color = 'blue')
ax3.set_xticklabels(['survived', "not survived"], rotation = 0)
plt.legend(['male / low class'], loc = 'best')

ax4 = fig.add_subplot(144, sharey = ax1)
train_data["Survived"][train_data["Sex"] == 'male'][train_data["Pclass"] == 3].value_counts().plot(kind = 'bar', color = 'steelblue')
ax4.set_xticklabels(['survived', "not survived"], rotation = 0)
#set_xlabel设置刻度，标签就是刻度；set_xticklabels，将其他值来当作刻度表标签
plt.legend(['male / high class'], loc = 'best')

fig = plt.figure()
fig.set(alpha = 0.2)
survived_0 = train_data["Embarked"][train_data["Survived"] == 0].value_counts()
print(type(survived_0))
survived_1 = train_data["Embarked"][train_data["Survived"] == 1].value_counts()
df = pd.DataFrame({"survived" : survived_1, "not survived" : survived_0})
df.plot(kind = 'bar', stacked = True)
plt.title("each embarked passengers survived distribution")
plt.ylabel("number of person")
plt.xlabel("embarked")
plt.show()

#train_data["Cabin"].value_counts()
fig = plt.figure()
fig.set(alpha = 0.2)

survived_0 = train_data["Cabin"][train_data["Survived"] == 0].value_counts()
survived_1 = train_data["Cabin"][train_data["Survived"] == 1].value_counts()
df = pd.DataFrame({"survived" : survived_1, "not survived" : survived_0})
df.plot(kind = 'bar', stacked = True)
plt.title("each Cabin passengers survived distributon")
plt.xlabel("Cabin")
plt.ylabel("number of person")
plt.show()

g = train_data.groupby(['SibSp', 'Survived'])#安照SibSp和Survived分类
df = pd.DataFrame(g.count()['PassengerId'])
print(df)

fig = plt.figure(figsize = (15, 5))
fig.set(alpha = 0.65)

ax1 = fig.add_subplot(131, sharey = ax1)
train_data["Embarked"][train_data["Pclass"] == 1].value_counts().plot(kind = 'bar', label = "first class", color = 'pink')
ax1.set_xticklabels(['S', 'C', 'Q'],rotation = 0)
plt.legend()

ax2 = fig.add_subplot(132, sharey = ax1)
train_data["Embarked"][train_data["Pclass"] == 2].value_counts().plot(kind = 'bar', label = "first class", color = 'red')
ax2.set_xticklabels(['S', 'C', 'Q'],rotation = 0)
plt.legend()

ax3 = fig.add_subplot(133, sharey = ax1)
train_data["Embarked"][train_data["Pclass"] == 1].value_counts().plot(kind = 'bar', label = "first class", color = 'blue')
ax3.set_xticklabels(['S', 'C', 'Q'],rotation = 0)
plt.legend()
plt.show()
```
